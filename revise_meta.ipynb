```python
import pandas as pd
```

```python
countries = pd.read_csv('https://raw.githubusercontent.com/miotomita/meta/main/countries_meta.csv')
```

追加

OWID

```python
sub = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/scripts/input/owid/subnational_population_2020.csv')
```

```python
sub = sub.loc[~sub.iso_code.isin(countries.code_OWID)].rename(columns={'location':'name_full','iso_code':'code_OWID','population':'population_OWID','source':'population_source'})
```

```python
sub['code_3digit'] = sub.code_OWID
sub['population_agg'] = sub.population_OWID
sub['name_short'] = sub.name_full
```

```python
sub['source'] = 'OWID'
```

```python
sub[['sovereignty_code','sovereignty_code_NK','region','sovereignty_name_NK','region_ja','category']]= countries.loc[countries.code_3digit=='GBR',['sovereignty_code','sovereignty_code_NK','region','sovereignty_name_NK','region_ja','category']].iloc[0].to_list()
```

```python
sub['name_ja'] = sub.code_OWID.map({'OWID_ENG':'イングランド', 'OWID_SCT':'スコットランド', 'OWID_WLS':'ウェールズ', 'OWID_NIR':'北アイルランド'})
```

```python
sub['notes'] = 'subnational area of GBR'
```

```python
sub
```

```python
countries = countries.append(sub).reset_index(drop=True)
```

```python
countries[['UN','WB','MOFA','EU','vac_data']] = False
```
